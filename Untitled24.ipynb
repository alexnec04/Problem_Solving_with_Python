```python
import itertools

def valid_arrangement(matrix):
    positions = [None] * 10
    for r in range(3):
        for c in range(3):
            num = matrix[r][c]
            positions[num] = (r, c)
    for k in range(1, 9):
        r1, c1 = positions[k]
        r2, c2 = positions[k + 1]
        if abs(r1 - r2) + abs(c1 - c2) != 1:
            return False
    return True

def count_valid_3x3_matrices():
    valid_count = 0
    for perm in itertools.permutations(range(1, 10)):
        matrix = [
            [perm[0], perm[1], perm[2]],
            [perm[3], perm[4], perm[5]],
            [perm[6], perm[7], perm[8]] ]
        if valid_arrangement(matrix):
            valid_count += 1
    return valid_count


print(count_valid_3x3_matrices())

```

```python

def valid_arrangement2(matrix):
    positions = [None] * 10
    for r in range(3):
        for c in range(3):
            positions[matrix[r][c]] = (r, c)
    for k in range(1, 9):
        r1, c1 = positions[k]
        r2, c2 = positions[k + 1]
        if not (abs(r1 - r2) <= 1 and abs(c1 - c2) <= 1):
            return False
    return True

def find_all_valid_matrices():
    valid_matrices = []
    for perm in itertools.permutations(range(1, 10)):
        matrix = [
            [perm[0], perm[1], perm[2]],
            [perm[3], perm[4], perm[5]],
            [perm[6], perm[7], perm[8]]]
        if valid_arrangement2(matrix):
            valid_matrices.append(matrix)
    return valid_matrices

def count_by_corner(valid_matrices):
    corner_counts = {i: 0 for i in range(1, 10)}
    corners = [(0, 0), (0, 2), (2, 0), (2, 2)]
    for matrix in valid_matrices:
        for r, c in corners:
            corner_counts[matrix[r][c]] += 1
    return corner_counts

def count_by_side_center(valid_matrices):
    side_center_counts = {i: 0 for i in range(1, 10)}
    side_centers = [(0, 1), (1, 0), (1, 2), (2, 1)]
    for matrix in valid_matrices:
        for r, c in side_centers:
            side_center_counts[matrix[r][c]] += 1
    return side_center_counts

def print_counts_table(title, counts):
    print(title)
    print("+--------+-------+")
    print("| Number | Count |")
    print("+--------+-------+")
    for num in range(1, 10):
        print(f"|   {num}    | {counts[num]:5d} |")
    print("+--------+-------+")
    print()

def count_by_center(valid_matrices):
    center_counts = {i: 0 for i in range(1, 10)}
    for matrix in valid_matrices:
        center_counts[matrix[1][1]] += 1
    return center_counts

def print_center_counts_table(center_counts):
    print("+--------+-------+")
    print("| Center | Count |")
    print("+--------+-------+")
    for num in range(1, 10):
        print(f"|   {num}    | {center_counts[num]:5d} |")
    print("+--------+-------+")


matrices = find_all_valid_matrices()
print(( len(matrices)))
corner_counts = count_by_corner(valid_matrices)
side_center_counts = count_by_side_center(valid_matrices)
print_counts_table("Corner Counts:", corner_counts)
print_counts_table("Side Center Counts:", side_center_counts)
center_counts = count_by_center(valid_matrices)
print("Counts by center number:")
print_center_counts_table(center_counts)
```
